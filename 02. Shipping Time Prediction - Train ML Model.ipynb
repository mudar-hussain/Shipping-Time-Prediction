# Shipping Time Prediction

## Machine Learning Model Training

## 1. Import Libraries

```python
import numpy as np
import pandas as pd
pd.set_option('display.max_columns', None)
```

## 2. Data Load

```python
path = r"Dataset/Preprocessed_SCMS_Delivery_History_Dataset.csv"
df = pd.read_csv(path)
```

```python
df.head()
```

```python
df.shape
```

```python
df.isna().sum().sum()
```

## 3. Split Data

```python
X = df.drop(['Diff in Delivery Date (Days)'], axis = 1)
y = df['Diff in Delivery Date (Days)']
```

```python
X
```

```python
y
```

```python
#Split data into training and test set for modelling
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 53)

print('Training data contains %.0f records'%len(X_train))
print('Test data contains %.0f records'%len(X_test))
```

## 4. Feature Scaling

```python
from sklearn.preprocessing import StandardScaler
```

```python
sc = StandardScaler()
```

```python
sc.fit(X_train)
X_train = sc.transform(X_train)
X_test = sc.transform(X_test)
```

```python
X_train
```

```python
y_train
```

## 5. Creating model for Delivery Date Prediction

```python
#For Model Evaluation
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score

import math
```

## A. Linear Regression

```python
from sklearn.linear_model import LinearRegression 

LR = LinearRegression()
LR.fit(X_train,y_train)

y_pred = LR.predict(X_test)
```

```python
LR.score(X_test, y_test)
```

```python
#Linear Regression Model Evaluation 

MAE = mean_absolute_error(y_test,y_pred)

MSE = mean_squared_error(y_test,y_pred)

RMSE = math.sqrt(MSE)

R2 = r2_score(y_test, y_pred) 

Ytest_Mean = y_test.mean() 

print("Mean Absolute Error:", MAE)
print("Mean Squared Error:", MSE)
print("Root Mean Squared Error:", RMSE)
print("R-Squared:", R2)
print("Test data mean:", Ytest_Mean)
print("RMSE/Mean:", RMSE/Ytest_Mean)
```

## B. Support Vector Regression

```python
from sklearn.svm import SVR

Regressor_SVR = SVR(kernel="rbf")
Regressor_SVR.fit(X_train,y_train.ravel())

y_pred_s = Regressor_SVR.predict(X_test)
```

```python
Regressor_SVR.score(X_test, y_test)
```

```python
#SVR Model Evaluation 

MAE_S = mean_absolute_error(y_test,y_pred_s)

MSE_S = mean_squared_error(y_test,y_pred_s) 

RMSE_S = math.sqrt(MSE_S)

R2_S = r2_score(y_test, y_pred_s) 

Ytest_Mean = y_test.mean()

print("Mean Absolute Error:", MAE_S)
print("Mean Squared Error:", MSE_S)
print("Root Mean Squared Error:", RMSE_S)
print("R-Squared:", R2_S)
print("Test data mean:", Ytest_Mean)
print("RMSE/Mean", RMSE_S/Ytest_Mean)
```

## C. Decision Tree Regression

```python
from sklearn.tree import DecisionTreeRegressor

regressor_dt = DecisionTreeRegressor()
regressor_dt.fit(X_train,y_train)

y_pred_dt = regressor_dt.predict(X_test)
```

```python
regressor_dt.score(X_test, y_test)
```

```python
#Decision Tree Regression Model Evaluation

MAE_DT = mean_absolute_error(y_test,y_pred_dt)

MSE_DT = mean_squared_error(y_test,y_pred_dt)

RMSE_DT = math.sqrt(MSE_DT) 

R2_DT = r2_score(y_test, y_pred_dt)

Ytest_Mean = y_test.mean()  

print("Mean Absolute Error:", MAE_DT)
print("Mean Squared Error:", MSE_DT)
print("Root Mean Squared Error:", RMSE_DT)
print("R-Squared:", R2_DT)
print("RMSE/Mean", RMSE_DT/Ytest_Mean)
```

## D. Random Forest Regression

```python
#HyperParameter Tuning - Check for best n_estimators
#1200 had the least RMSE

from sklearn.ensemble import RandomForestRegressor

RF_RMSE = []
estimators = [100,200,300,400,500,600,700,800,900,1000,1200,1500]
for i in estimators:
    RFR = RandomForestRegressor(n_estimators = i)
    RFR.fit(X_train,y_train.ravel())
    y_pred_rf = RFR.predict(X_test)
    MSE_RF = mean_squared_error(y_test,y_pred_rf)
    RMSE_RF = math.sqrt(MSE_RF) 
    print(str(i) + " -> " + str(RMSE_RF))
    RF_RMSE.append(RMSE_RF)
    

print(RF_RMSE)
```

```python
#Build model with n_estimators = 1200
 
from sklearn.ensemble import RandomForestRegressor 

regressor_rf = RandomForestRegressor(n_estimators = 1200)
regressor_rf.fit(X_train,y_train.ravel());

y_pred_rf = regressor_rf.predict(X_test)
```

```python
regressor_rf.score(X_test, y_test)
```

```python
#Random Forest Regression Evaluation

MAE_RF = mean_absolute_error(y_test,y_pred_rf) 

MSE_RF = mean_squared_error(y_test,y_pred_rf) 

RMSE_RF = math.sqrt(MSE_RF) 

R2_RF = r2_score(y_test, y_pred_rf) 

Ytest_Mean = y_test.mean() 

print("Mean Absolute Error:", MAE_RF)
print("Mean Squared Error:", MSE_RF)
print("Root Mean Squared Error:", RMSE_RF)
print("R-Squared:", R2_RF)
print("RMSE/Mean", RMSE_RF/Ytest_Mean) 
```

## E. K-Nearest Neighbour (KNN) Regression

```python
#HyperParameter Tuning - Check for best n_neighbors
#35 had the least RMSE

from sklearn.neighbors import KNeighborsRegressor

KNN_RMSE = []
for i in range(1,50):
    KNNR = KNeighborsRegressor(n_neighbors = i)
    KNNR.fit(X_train,y_train)
    y_pred_knn = KNNR.predict(X_test)
    MSE_KNN = mean_squared_error(y_test,y_pred_knn)
    RMSE_KNN = math.sqrt(MSE_KNN) 
    print(str(i) + " -> " + str(RMSE_KNN))
    KNN_RMSE.append(RMSE_KNN)
    

print(KNN_RMSE)
```

```python
#Build model with n_neighbors = 35
from sklearn.neighbors import KNeighborsRegressor
regressor_knn = KNeighborsRegressor(n_neighbors = 35)
 
regressor_knn.fit(X_train,y_train)


y_pred_knn = regressor_knn.predict(X_test)

#y_pred_knn = y_pred_knn.reshape(-1,1)
#y_pred_knn = Sc2.inverse_transform(y_pred_knn)
```

```python
regressor_knn.score(X_test, y_test)
```

```python
#KNN Regression Model Evaluation

MAE_KNN = mean_absolute_error(y_test,y_pred_knn) 

MSE_KNN = mean_squared_error(y_test,y_pred_knn) 

RMSE_KNN = math.sqrt(MSE_KNN) 

R2_KNN = r2_score(y_test, y_pred_knn) 

Ytest_Mean = y_test.mean()  


print("Mean Absolute Error:", MAE_KNN)
print("Mean Squared Error:", MSE_KNN)
print("Root Mean Squared Error:", RMSE_KNN)
print("R-Squared:", R2_KNN)
print("RMSE/Mean", RMSE_KNN/Ytest_Mean)
```

# 6. Conclusion

### Random Forest Regression performed the best with the lowest RMSE, so we will predict delivery time for the test data using the random forest regression model.
